```python
varlist =  ['fueltype']

# Defining the map function
def binary_map(x):
    return x.map({'gas': 1, "diesel": 0})

# Applying the function to the housing list
df[varlist] = df[varlist].apply(binary_map)
```

```python
varlist =  ['doornumber']

# Defining the map function
def binary_map(x):
    return x.map({'four': 1, "two": 0})

# Applying the function to the housing list
df[varlist] = df[varlist].apply(binary_map)
```

```python
varlist =  ['enginelocation']

# Defining the map function
def binary_map(x):
    return x.map({'front': 1, "rear": 0})

# Applying the function to the housing list
df[varlist] = df[varlist].apply(binary_map)
```

```python
del df["aspiration"]
del df["carbody"]
del df["drivewheel"]
del df["CarName"]
del df['symboling']
del df['car_ID']
del df["enginetype"]
del df['cylindernumber']
del df['fuelsystem']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
num_vars = ['wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight','enginesize',"boreratio","stroke","compressionratio","horsepower","peakrpm","citympg","highwaympg","price"]

df[num_vars] = scaler.fit_transform(df[num_vars])
```

```python
x = df.drop(columns = "price")
y = df['price']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split # here we have to import library to split the parameter
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)# here we have to split it into tranning and testing model
```

```python

import statsmodels.api as sm
x_train_lm = sm.add_constant(x_train)

lr_1 = sm.OLS(y_train, x_train_lm).fit()

lr_1.params
```

```python
print(lr_1.summary())
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train, y_train)

rfe = RFE(lm, n_features_to_select=5)             # running RFE
rfe = rfe.fit(x_train, y_train)
```

```python
list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = x_train.columns[rfe.support_]
col
```

```python
X_train_rfe = x_train[col]
```

```python
import statsmodels.api as sm
x_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,x_train_rfe).fit()   # Running the linear model
```

```python
print(lm.summary())
```

```python
from sklearn.linear_model import LinearRegression # here we have to import library
Model=LinearRegression() #here we have trained the model
Model.fit(x_train_rfe,y_train)
```

```python
y_pred=Model.predict(x_train_rfe)
print(y_pred)
```

```python
Model.score(x_train_rfe,y_train)
```

```python
from sklearn.metrics import mean_squared_error
mse_train=mean_squared_error(y_train,y_pred)
mse_train
```
